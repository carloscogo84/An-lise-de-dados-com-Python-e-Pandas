Com base nos dados acima criaremos novas colunas no dataset com base nas colunas existentes

```python
# Verificando a linha 1 do dataframe
df.head(1)
```

```python
# Tempo de envio da compra
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"] 
df
```

```python
# Custo total do produto
# Criando coluna Custo Total
df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head()
```

```python
# Cuato Total, arredondando valor
round(df["Custo Total"].sum(), 2)
```

```python
# Com a receita e custo e o total, podemos achar o Lucro total
# Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"]  = df["Valor Venda"] - df["Custo Total"] 
df.head()
```

```python
# Lucro líquido da venda
df["Lucro Líquido"] = df["lucro"] - df["Valor Desconto"]
df.head()
```

```python
# Percentual do Lucro bruto da venda
df["% Lucro Bruto"] = (1 - (df["Custo Total"] / df["Valor Venda"])) * 100
df.head()
```

```python
# Percentual do Lucro líquido da venda
df["% Lucro Líquido"] = (1 - (df["Custo Total"] / (df["Valor Venda"] - df["Valor Desconto"]))) * 100
df.head()
```

```python
total_vendas = df["Valor Venda"].sum()
total_custos = df["Custo Total"].sum()
total_descontos = df["Valor Desconto"].sum()
total_lucro_bruto = df["lucro"].sum()
total_lucro_liquido = df["Lucro Líquido"].sum()
```

```python
# Formatando valores com a biblioteca locale
import locale
locale.setlocale(locale.LC_MONETARY, 'en_US.UTF-8')

def valores():
    print("Total de vendas: ", locale.currency(total_vendas))
    print("Total de custos: ", locale.currency(total_custos))
    print("Total de descontos", locale.currency(total_descontos))
    print("Lucro bruto", locale.currency(total_lucro_bruto))
    print("Lucro líquido", locale.currency(total_lucro_liquido))

valores()
```

## Análise Gráfica dos Dados
Com os gráficos para melhor visualizac1ão, fica mais claros os insights gerados. Vamos gerar gráficos dessas explorações dos dados.

```python
# Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
# Identificando o tempo de entrega
x = df["Tempo Envio"].value_counts().values
y = df["Tempo Envio"].value_counts().index
sns.barplot(x=x, y=y)
plt.show()
```

```python
# Selecionando apenas as vendas de 2008 e 2009
df_2008 = df[df["Data Venda"].dt.year == 2008]
df_2009 = df[df["Data Venda"].dt.year == 2009]

# Comparativo de lucro bruto por mês
df_2008.groupby(df_2008["Data Venda"].dt.month)["lucro"].sum().plot(color="green", label="Ano 2008")
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(color="black", label="Ano 2009")
plt.title("Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro Bruto")
plt.legend();
```
